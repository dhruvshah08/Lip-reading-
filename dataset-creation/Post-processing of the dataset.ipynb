```python
#Transliterate
import os
from ai4bharat.transliteration import XlitEngine
dir_name = "D:\\THEDATASET"
dirs = os.listdir(dir_name)
e = XlitEngine(src_script_type="indic", beam_width=10, rescore=False)
for folder in dirs:
    name = dir_name+'\\'+folder
    folder_id = int(folder)
    if folder_id <=9 and folder_id>=5:
    
        print('Now processing: ',folder_id)
        for item in os.listdir(name):
            try:
                if item.endswith(".txt"):
                    with open(name+ "\\"+item, 'r+', encoding="utf-8") as f:   
                        file = f.readlines()
                        line1 = file[0]
                        textToTransliterate1 = line1[6:]
                        out = e.translit_sentence(textToTransliterate1,lang_code="hi")
                        transliteratedText = line1[:6] + out + "\n"
                        line2 = file[1]
                        line3 = file[2]
                        line4 = file[3]
                        f1 = [transliteratedText,line2,line3,line4]
                        
                        for i in range(4,len(file)):
                            line = file[i]
                            textToTransliterate2 = line.split(' ')[0] 
                            out1 = e.translit_sentence(textToTransliterate2,lang_code="hi")
                            len1 = len(textToTransliterate2)
                            finalLine =  out1 + line[len1:]
                            f1.append(finalLine)
                        f.seek(0)
                        f.truncate(0)
                        f.writelines(f1)
                        f.close()
            except:
                continue
            
```
