```python
from pytube import YouTube
def download_video(vid_id):
    link = 'https://www.youtube.com/watch?v=' + vid_id
    youtubeObject = YouTube(link)
    youtubeObject = youtubeObject.streams.get_highest_resolution()
    try:
        youtubeObject.download()
    except:
        print("An error has occurred")
    print("Download is completed successfully")
download_video(video_id)
```

```python
import os
import sys
import wave
import json
```

```python
#Clip based on each video
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip
import moviepy.editor as mp
import moviepy.editor as mp
from pydub import AudioSegment
import os
from google.cloud import speech

os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = 'lip-reading-project-374705-89169d76c96a.json'
speech_client = speech.SpeechClient()

for i in range(length):
    text = list_of_translated_script[i]['text']
    start = float(list_of_translated_script[i]['start'])
    duration = float(list_of_translated_script[i]['duration'])
    end = float(start + duration)
    file = str(i).zfill(5)
    file_name = file +'.mp4'
    audio_filename = file + ".wav"
    text_file_name = file + ".txt"
    #clip
    ffmpeg_extract_subclip("video1.mp4", start, end, targetname=file_name)
    
    #Convert .mp4 to .wav
    clip = mp.VideoFileClip(file_name)
    
    clip.audio.write_audiofile(audio_filename)
    
    #Convert to one-channel .wav file
    sound = AudioSegment.from_wav(audio_filename)
    sound = sound.set_channels(1)
    sound.export(audio_filename, format="wav")
 
    #FACE RECOG REMOVE NOT DHRUV RATHEE
    
    if not os.path.exists(audio_filename):
        print(f"File '{audio_filename}' doesn't exist")
        sys.exit()

    print(f"Reading your file '{audio_filename}'...")
    wf = wave.open(audio_filename, "rb")
    print(f"'{audio_filename}' file was successfully read")
    
    if wf.getnchannels() != 1 or wf.getsampwidth() != 2 or wf.getcomptype() != "NONE":
        print("Audio file must be WAV format mono PCM.")
        sys.exit()
        
# ------------------------------------------------------------------------------------------
    media_file_name_wav = audio_filename

    with open(media_file_name_wav, 'rb') as f2:
        byte_data_wav = f2.read()
    audio_wav = speech.RecognitionAudio(content=byte_data_wav)

    config_wav = speech.RecognitionConfig(
        sample_rate_hertz=44100,
        language_code='hi',
        audio_channel_count=1,
        enable_word_time_offsets = True
    )
    response_standard_wav = speech_client.recognize(
        config=config_wav,
        audio=audio_wav
    )

    for result in response_standard_wav.results: 
        alternative = result.alternatives[0]
        list_of_row_details = []
        text = alternative.transcript
        for word_info in alternative.words: 
            word = word_info.word
            start_time = word_info.start_time 
            start_time = start_time.seconds + start_time.microseconds  * 1e-6
            end_time = word_info.end_time 
            end_time = end_time.seconds + end_time.microseconds  * 1e-6
            list_of_row_details.append(word + ' ' + str(start_time) +  ' ' + str(end_time) + ' ' + str(0.0))
    file = open(text_file_name,'w', encoding="utf-8")
    file.write('Text: '+text)
    file.write('\n')
    file.write('Conf: '+ '0')
    file.write('\n\n')
    file.write('WORD START END ASDSCORE')
    for row in list_of_row_details:
        file.write("\n"+row)
    file.close()
    break
```
